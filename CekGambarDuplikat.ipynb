# Cek

```python
import os
import hashlib
from PIL import Image
import matplotlib.pyplot as plt

def file_hash(filepath):
    """Menghitung MD5 hash dari isi file."""
    with open(filepath, 'rb') as f:
        file_hash = hashlib.md5()
        while chunk := f.read(8192):
            file_hash.update(chunk)
    return file_hash.hexdigest()

def check_duplicate_images(folder_path):
    # Dictionary untuk menyimpan hash dan nama file
    hashes = {}
    duplicates = []

    # Iterasi melalui semua file di folder
    for filename in os.listdir(folder_path):
        # Dapatkan path penuh file
        file_path = os.path.join(folder_path, filename)
        
        # Buka dan hitung hash gambar
        if os.path.isfile(file_path): # Pastikan itu file, bukan folder
            try:
                # Menghitung hash
                img_hash = file_hash(file_path)
                
                # Cek apakah hash sudah ada di dictionary
                if img_hash in hashes:
                    duplicates.append((file_path, os.path.join(folder_path, hashes[img_hash])))
                else:
                    hashes[img_hash] = filename
            except Exception as e:
                print(f"Error processing file {filename}: {e}")
    
    # Menampilkan gambar yang duplicate
    if duplicates:
        print("Found duplicates:")
        for dup in duplicates:
            print(f"{dup[0]} is a duplicate of {dup[1]}")
            img1 = Image.open(dup[0])
            img2 = Image.open(dup[1])
            plt.figure(figsize=(10,5))
            plt.subplot(1,2,1)
            plt.imshow(img1)
            plt.axis('off')
            plt.title("Duplicate 1")
            plt.subplot(1,2,2)
            plt.imshow(img2)
            plt.axis('off')
            plt.title("Duplicate 2")
            plt.show()
    else:
        print("No duplicate files found.")

# Path folder yang berisi gambar
folder_path = "Gym Equipment Mentah/Stair Climber"

# Jalankan fungsi
check_duplicate_images(folder_path)
```

# Delete

```python
import os
import hashlib
from PIL import Image

def file_hash(filepath):
    """Menghitung MD5 hash dari isi file."""
    with open(filepath, 'rb') as f:
        file_hash = hashlib.md5()
        while chunk := f.read(8192):
            file_hash.update(chunk)
    return file_hash.hexdigest()

def check_and_delete_duplicate_images(folder_path):
    # Dictionary untuk menyimpan hash dan nama file
    hashes = {}
    duplicates = []

    # Iterasi melalui semua file di folder
    for filename in os.listdir(folder_path):
        # Dapatkan path penuh file
        file_path = os.path.join(folder_path, filename)
        
        # Buka dan hitung hash gambar
        if os.path.isfile(file_path): # Pastikan itu file, bukan folder
            try:
                # Menghitung hash
                img_hash = file_hash(file_path)
                
                # Cek apakah hash sudah ada di dictionary
                if img_hash in hashes:
                    duplicates.append((file_path, os.path.join(folder_path, hashes[img_hash])))
                else:
                    hashes[img_hash] = filename
            except Exception as e:
                print(f"Error processing file {filename}: {e}")
    
    # Cetak dan hapus file yang duplicate
    if duplicates:
        print("Found duplicates, deleting:")
        for dup in duplicates:
            print(f"Deleting {dup[1]} which is a duplicate of {dup[0]}")
            os.remove(dup[1])  # Hapus file duplikat
    else:
        print("No duplicate files found.")

# Path folder yang berisi gambar
folder_path = "Gym Equipment Mentah/Stair Climber"

# Jalankan fungsi
check_and_delete_duplicate_images(folder_path)
```
